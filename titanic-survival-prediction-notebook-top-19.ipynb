**TRAINING DATA**

```python
X = df.iloc[:,1:].values
y = df.iloc[:, 0].values
```

```python
X_test = df1.iloc[:,:].values
```

**ENCODING CATEGORICAL VARIABLES**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X_test = np.array(ct.fit_transform(X_test))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1])], remainder='passthrough')
X_test = np.array(ct.fit_transform(X_test))
```

```python
import numpy as np
np.set_printoptions(threshold=np.inf)
```

**TRAINING AND FITTING THE DATA Using Decision tree classifier**

```python
from sklearn.tree import DecisionTreeClassifier
d = DecisionTreeClassifier(max_depth = 5)
d.fit(X,y)
```

**TRAINING AND FITTING THE DATA Using KNN(K-nearest neighbour)**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X, y)
```

**Predicting results for KNN**

```python
y_kpred = classifier.predict(X_test)
```
